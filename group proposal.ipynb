```r
library(tidyverse)
placement_data <- read_csv("Placement_Data_Full_Class.csv")
placement_data 
```

```r
new_placement_data <-placement_data|>
select(ssc_p, hsc_p, degree_p, etest_p, mba_p, status)
new_placement_data
```

```r
clean_placement_data <-new_placement_data|>
pivot_longer(cols= ssc_p:mba_p,
             names_to="level_of_education",
            values_to= "grades_percentage")
clean_placement_data

```

```r
mean_placement_data <-clean_placement_data|>
group_by(status, level_of_education)|>
summarize(grades_percentage=mean(grades_percentage,na.rm=TRUE))
mean_placement_data
```

```r

plot_placement_data <- mean_placement_data |>
  ggplot(aes(x = level_of_education, y = grades_percentage, fill = status)) +
  geom_bar(stat = "identity", position="dodge") +
  labs(x = "Level of Education", y = "Grade Percentage", fill = "Status") +
  theme(text = element_text(size = 20))

plot_placement_data
```
